```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def func(y): return y ** 2

def der_func(y): return 2 * y

# Create points for the function
x = np.arange(-100, 100, 0.1)
y = func(x)


current_value = (99, func(99))


plt.plot(x, y)
plt.scatter(current_value[0], current_value[1], color="red")
plt.show()

```

```python
learning_rate = 0.01
error = 0.025

def gradient_descent_step(x, lr):
    return x - lr * der_func(x)

current_x = current_value[0]
current_y = current_value[1]

while current_y > error:
    new_x = gradient_descent_step(current_x, learning_rate)
    new_y = func(new_x)

    current_x = new_x
    current_y = new_y

    plt.plot(x, y)
    plt.scatter(new_x, new_y, color="red")
    plt.show()
    plt.pause(0.01)
    plt.clf()
```
